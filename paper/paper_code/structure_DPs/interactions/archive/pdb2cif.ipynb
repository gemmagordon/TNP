Convert PDB files to mmcif to set up for arpeggio (interactions analysis)

```python
import os
from Bio.PDB import PDBParser, MMCIFIO
from pathlib import Path
#from multiprocessing import Pool
```

```python
#cpus = 80

dataset_name = 'vh_oas'
structure_path = '/data/localhost/gordon/TNP_Project/results/debugged_nbb2_results/new_'+dataset_name+'_output/Raw_Model_Outputs/'
structures = [str(s).split('/')[-1] for s in list(Path(structure_path).glob('**/*.pdb'))]

output_path = '/vols/opig/users/gordon/tnp_project/greiff_results/'+dataset_name+'_cif/'
```

```python
def pdb2cif(pdb, structure_path = structure_path):
    try:
        parser = PDBParser(QUIET=True)
        structure = parser.get_structure('', structure_path + pdb[:-4] + '/' + pdb) 
        print(structure)
        io = MMCIFIO()
        io.set_structure(structure)
        io.save("{}.cif".format(output_path + pdb[:-4]))
        print('{} done'.format(pdb[:-4]))
    except Exception:
        print('{} failed'.format(pdb))
```

```python
# if __name__ == '__main__':
#     pool = Pool(processes = cpus)
#     results = pool.map(pdb2cif, structures)
#     pool.close()
#     pool.join()

for structure in structures:
    pdb2cif(structure, structure_path=structure_path)
```
